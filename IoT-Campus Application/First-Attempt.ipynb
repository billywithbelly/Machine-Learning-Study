```python
df = pd.read_csv('./parsed.csv', parse_dates=True, dayfirst=True)
df.columns = ['id', 'date', 'hour', 'flag', 'raw_data']
df.head(10)
```

```python
sns.set(style='dark', color_codes=True)
sns.set_context("poster")
#df.plot(kind="scatter", x="id", y="hour")
#sns.jointplot(x="id", y="hour", data=df, size=5)
```

```python
#ax=sns.boxplot(x="id", y="hour", data=df)
ax=sns.stripplot(x="id", y="hour", data=df, jitter=True, edgecolor="gray")
#ax.legend(fontsize=15)
sns.plt.title('id/hour')
```

```python
ax = sns.violinplot(x="id", y="hour", data=df, size=1, hue="id", aspect=.7)
sns.plt.title('id/hour')
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
#plt.savefig('./violinplot-id-hour.png', dpi=300)
```

```python
ax = sns.violinplot(x="date", y="hour", data=df, size=5, title="date-hour", hue='date')
ax.legend(fontsize=5)
ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
ax.set_xticklabels(ax.get_xticklabels(), rotation=30)
sns.plt.title('date/hour')
#sns.stripplot(x="date", y="hour", data=df, jitter=True, edgecolor="gray", size=5)
plt.savefig('./violinplot-date-hour.png', dpi=300)
```
